# Using sklearn digital daataset to make classification using random forest

Digits dataset from sklearn

```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.gray() 
for i in range(4):
    plt.matshow(digits.images[i]) 
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

so the output tells that in row0 all these 64 data will map to zero

### Train and the model and prediction

```python
X = df.drop('target',axis='columns')
y = df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier   #we use ensemble as we are using diff algm to train model
model = RandomForestClassifier(n_estimators=20)   #it uses 20 random trees and using nestimator increases accuracy
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)    
```

# Confusion Matrix

```python
y_predicted = model.predict(X_test)

#y preedicted is predicted by model, y_test is the truth/real value
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
import pandas as pd
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'],axis='columns'),iris.target,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
model = RandomForestClassifier(n_estimators=10)
model.fit(X_train, y_train)
model.score(X_test,y_test)
```
